```python
from IPython.display import clear_output
from matplotlib.gridspec import GridSpec
import matplotlib.pyplot as plt
import numpy as np
import sys
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision.utils import make_grid

sys.path.append('..')
from utils import (
    ImSet,
    ConditionalVariationalAutoEncoder,
    ConditionalDiscriminator,
    show_tensor
)
%matplotlib inline
```

# Joint CVAE / CGAN

- I want the CGAN generator to also function as a CVAE decoder
- The discriminator can remain separate from the CVAE encoder, though I'll initialize

```python
device = torch.device(0)

imset = ImSet()

batch_size = len(imset) // 5
loader = DataLoader(imset, shuffle=True, batch_size=batch_size, drop_last=True)

cvae = ConditionalVariationalAutoEncoder().to(device)

n_trials = 1000
n_epochs = 10000

model_fname = 'cvae.pth'

latent_size = 4096
```

```python
dsc = ConditionalDiscriminator().to(device)
```

```python
lossfun = nn.BCELoss()

# create labels
fake_labels = torch.zeros(batch_size, 1).to(device)
real_labels = torch.ones(batch_size, 1).to(device)
```

```python
with open('cvaegan_log.txt', 'w') as fout:
    fout.write('trial, epoch, ae_loss, d_loss, g_loss, p_real, p_fake\n')
```

```python
for trial in range(n_trials):
    
    d_losses = []
    g_losses = []
    ae_losses = []

    real_preds = []
    fake_preds = []
    
    # load the pretrained CVAE 
    # cvae.load_state_dict(torch.load(model_fname), strict=False)
    
    # extract generator
    gen = cvae.dec
    
    # load encoder weights to the discriminator
    # dsc.load_state_dict(cvae.enc.state_dict(), strict=False)
    
    # initialize optimizers
    opt_ae = torch.optim.Adam(cvae.parameters(), lr=.0002, betas=(.5,.999))
    opt_d = torch.optim.Adam(dsc.parameters(), lr=.0002, betas=(.5,.999))
    opt_g = torch.optim.Adam(gen.parameters(), lr=.0002, betas=(.5,.999))

    # loop through epochs
    for epoch in range(n_epochs):
        # loop through batches
        for (img_fname, img, species_, class_, gender_) in loader:
            
            # transfer to GPU
            img = img.to(device)
            species_ = species_.to(device)
            class_ = class_.to(device)
            gender_ = gender_.to(device)
            
            # train autoencoder
            # =================
            
            img_t = cvae(img, species_, class_, gender_)
            loss = ((img - img_t)**2).sum() + cvae.enc.kl

            opt_ae.zero_grad()
            loss.backward()
            opt_ae.step()

            ae_losses.append(loss.item())
            
            # train discriminator
            # ===================
            
            # forward pass for real
            pred_real = dsc(img, species_, class_, gender_)
            
            d_loss_real = lossfun(pred_real, real_labels)
            
            # forward pass for fake
            fake_data = torch.randn(batch_size, latent_size).to(device)
            fake_images = gen(fake_data, species_, class_, gender_)
            pred_fake = dsc(fake_images, species_, class_, gender_)
            d_loss_fake = lossfun(pred_fake, fake_labels)
            
            # backprop
            d_loss = d_loss_real + d_loss_fake
            opt_d.zero_grad()
            d_loss.backward()
            opt_d.step()
            d_losses.append(d_loss.item())
            
            # train generator
            # ===============
            
            fake_data = torch.randn(batch_size, latent_size).to(device)
            fake_images = gen(fake_data, species_, class_, gender_)
            pred_fake = dsc(fake_images, species_, class_, gender_)
            
            # backprop
            g_loss = lossfun(pred_fake, real_labels)
            opt_g.zero_grad()
            g_loss.backward()
            opt_g.step()
            g_losses.append(g_loss.item())
            
            real_preds.append(torch.mean((pred_real > 0.5).float()).detach().cpu())
            fake_preds.append(torch.mean((pred_fake > 0.5).float()).detach().cpu())
            
            d_losses = d_losses[:1000]
            g_losses = g_losses[:1000]
            ae_losses = ae_losses[:1000]
            
            with open('cvaegan_log.txt', 'a') as fout:
                #fout.write('trial,epoch,ae_loss,d_loss,g_loss,p_real,p_fake\n')
                fout.write(f'{trial}, {epoch}, {ae_losses[-1]:.0f}, {d_losses[-1]:.2f}, {g_losses[-1]:.2f}, {real_preds[-1]:.2f}, {fake_preds[-1]:.2f}\n')
            
            # visualize output
            # ================
            
            # show fake images
            grid = make_grid(fake_images[:36], nrow=12)

            # show reconstructed images
            ae_grid = make_grid(torch.concat((img[:12], img_t[:12])), nrow=12)
            
            fig = plt.figure(figsize=(15, 12))
            gs = GridSpec(8, 1, figure=fig)
            ax0 = fig.add_subplot(gs[:3])
            ax1 = fig.add_subplot(gs[3])
            ax2 = fig.add_subplot(gs[4])
            ax3 = fig.add_subplot(gs[5])
            ax4 = fig.add_subplot(gs[6:])

            clear_output(wait=True)
            show_tensor(grid, ax0)
            ax1.plot(g_losses)
            ax1.plot(d_losses)
            ax2.plot(real_preds)
            ax2.plot(fake_preds)
            ax3.plot(ae_losses)
            show_tensor(ae_grid, ax4)
            plt.show()
    
        if epoch % 10 == 0:
            fig.savefig(f'CGAN_output/trial_{trial:09}_epoch_{epoch:09}.png')
        if epoch % 100 == 0:
            torch.save(gen.state_dict(), f'CGAN_output/trial_{trial:09}_epoch_{epoch:09}_gen.pth')
            torch.save(dsc.state_dict(), f'CGAN_output/trial_{trial:09}_epoch_{epoch:09}_dsc.pth')
```
